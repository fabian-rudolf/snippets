```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

y = [720, 360, 180, 147, 108, 131, 86, 88, 69, 70, 105, 80, 61, 71]
X = list(np.arange(0, len(y)))

plt.scatter(X,y,c=y,cmap='Greens_r')

ax.set_facecolor('#eee')

ax.set_title(r'solving $\int \frac{x^2}{\sqrt{5-2x^3}} \,dx $ by hand via substitution')
plt.ylabel('time required [seconds]')
plt.xlabel('attempt number')

ax.annotate(f'n = {sum(y)}',
            xy=(1, 1), xycoords='data',
            xytext=(0.98, 0.95), textcoords='axes fraction',
            horizontalalignment='right', verticalalignment='top')

plt.show()

fig.savefig('repitition.png',bbox_inches='tight')
```
