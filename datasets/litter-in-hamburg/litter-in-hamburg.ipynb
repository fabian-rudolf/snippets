# litter at Hamburg Fischmarkt 29 May 2021

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
X = ['single-use plastics', 'single-use paper', 'returnable glass beer bottles', 'pieces of broken glass', 'aluminum', 'medical face masks', 'food leftovers', 'styrofoam', 'ceramics', 'composite waste']
y = [ 24, 22, 16, 10, 9, 8, 3, 3, 1, 1]

ax.bar(X,y,color=(0.2, 0.4, 0.6, 0.6))

ax.set_title('type of litter at Fischmarkt\n Hamburg 29 May 2021', fontsize='large')
plt.ylabel('pieces of litter', fontsize='large')

plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='large'  
)

ax.annotate(f'n = {sum(y)}',
            xy=(1, 1), xycoords='data',
            xytext=(0.98, 0.95), textcoords='axes fraction',
            horizontalalignment='right', verticalalignment='top')

plt.show()

fig.savefig('trash-fischmarkt-2021-05-29.jpg',bbox_inches='tight')
```
